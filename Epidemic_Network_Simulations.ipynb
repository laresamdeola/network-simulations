```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
def analyze_network(G):
    # Degree Distribution
    degrees = [G.degree(n) for n in G.nodes()]
    plt.hist(degrees, bins=range(min(degrees), max(degrees) + 1), align='left', color='skyblue')
    plt.title("Degree Distribution")
    plt.xlabel("Degree (Number of connections)")
    plt.ylabel("Frequency")
    plt.show()

    # Average Clustering Coefficient
    clustering_coeff = nx.average_clustering(G)
    print(f"Average Clustering Coefficient: {clustering_coeff:.4f}")

    # Local Clustering Coefficient - Gives individual nodes clustering coefificent
    # local_clustering_coeff = nx.clustering(G)
    # print(f"Local Clustering Coefficient: {local_clustering_coeff}")

    # Global Clustering Coefficient
    global_clustering_coeff = nx.transitivity(G)
    print(f"Global Clustering Coefficient: {global_clustering_coeff:.4f}")

    # Connected Components
    connected_components = nx.number_connected_components(G)
    print(f"Number of Connected Components: {connected_components}")
```

Epidemic Spread

```python
def create_epidemic_network(n, infection_probability):
  '''
  - n: number of individuals (nodes)
  - infection_probability: probability of infection spreading between individuals
  '''
  G = nx.Graph()

  for i in range(n):
    G.add_node(i)
    for j in range(i+1, n):
      if random.random() < infection_probability:
        G.add_edge(i, j)

  return G
```

```python
G_epidemic = create_epidemic_network(20, 0.1)
nx.draw(G_epidemic, with_labels=True, node_color='red', edge_color='gray')
plt.show()
```

```python
analyze_network(G_epidemic)
```
